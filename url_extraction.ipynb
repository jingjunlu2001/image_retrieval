```python
# !pip install urlextract
import codecs
import re
from urlextract import URLExtract
```

```python
f = codecs.open("view-source_https___theporndude.com.html", 'r', 'utf-8')
```

```python
html = f.read()
```

```python
%%time
# Method 1
extractor = URLExtract()
urls = extractor.find_urls(html)
print('URLs found: ' + str(len(urls)))
print(urls)
```

```python
%%time
# Method 2
urls = re.findall("(https?://[^\s\'\"<]+)", html)
print('URLs found: ' + str(len(urls)))
print(urls)
```
